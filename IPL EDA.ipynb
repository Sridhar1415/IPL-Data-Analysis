# Win % of each team

```python
wins_des=pd.merge(matches_played_teams,Total_wins_each_team,how='inner',on='Teams').sort_values(by='no_of_matches',ascending=False)
```

```python
wins_des
```

```python
wins_des['%win']=(wins_des['no_of_wins']/wins_des['no_of_matches'])*100
```

```python
wins_des=wins_des.sort_values(by='%win',ascending=False)
```

```python
fig=plt.figure(figsize=(6,6))
axes=fig.add_axes([0,0,1,1])
axes.set_xticklabels(labels=wins_des['Teams'],rotation=90)
sns.barplot(x='Teams',y='%win',data=wins_des)
```

# Lucky venues for a team

```python
def lucky(matches,team_name):
    return matches[matches['winner']==team_name]['venue'].value_counts().nlargest(10)
```

```python
lucky_venues=lucky(matches,'Mumbai Indians').reset_index().rename(columns={'index':'Venue','venue':'no_of_matches_won'})
```

```python
lucky_venues
```

```python
fig=plt.figure(figsize=(6,6))
axes=fig.add_axes([0,0,1,1])
plt.pie(x='no_of_matches_won',labels='Venue',data=lucky_venues,autopct='%1.2f%%')
```
